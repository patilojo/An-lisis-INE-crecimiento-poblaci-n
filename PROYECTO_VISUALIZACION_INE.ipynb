### ¿Cómo ha ido variando la población española a lo largo del tiempo?

```python
df.fecha[df.dato_segmentado == 'Total']
```

```python
# tradicionalmente la serie de las x pinta la parte temporal 
```

```python
# tenemos 3 series de datos , entonces necesitamos acotar , en total o mujeres y hombres
```

```python
plt.figure(figsize = (9,4))
plt.plot(df.fecha[df.dato_segmentado == 'Total'], df.poblacion[df.dato_segmentado == 'Total']/1000000, marker= 'o')
plt.xlabel('Año')
plt.ylabel('Población (millones)')
plt.title('Población Total', pad = 20);
```

```python
# una ve vista la grafica vemos que los datos son coherentes, que la media se cruza solo una vez
```

```python
# en estos años de abajo tenemos incrementos poblacionales negativos y en la grafica se ve como en esos años baja un poquito, por lo tanto es coherente
# tb vemmos que tenemos diferentes ritmos de creciemtno
```
